# q1_delay_rate_by_region

```sql
%sql
SELECT
  region,
  COUNT(*) AS num_deliveries,
  ROUND(AVG(delayed) * 100, 1) AS delay_rate_percent,
  ROUND(AVG(delivery_time_hours), 2) AS avg_delivery_time_hours,
  ROUND(AVG(distance_km), 2) AS avg_distance_km
FROM workspace.delivery_logistics.delivery_logistic_clean
GROUP BY region
ORDER BY delay_rate_percentÂ DESC;
```

#q2_weather_vs_delay

```sql
SELECT
  weather_condition,
  COUNT(*) AS num_deliveries,
  ROUND(AVG(delayed) * 100, 1) AS delay_rate_percent,
  ROUND(AVG(delay_hours), 2) AS avg_delay_hours
FROM workspace.delivery_logistics.delivery_logistic_clean
GROUP BY weather_condition
ORDER BY delay_rate_percent DESC;
```

#q3_cost_efficiency_by_mode

```sql
SELECT
  delivery_mode,
  COUNT(*) AS num_deliveries,
  ROUND(AVG(delivery_cost), 2) AS avg_delivery_cost,
  ROUND(AVG(distance_km), 2) AS avg_distance_km,
  ROUND(AVG(cost_per_km), 2) AS avg_cost_per_km
FROM workspace.delivery_logistics.delivery_logistic_clean
GROUP BY delivery_mode
ORDER BY avg_cost_per_km;
```

# q4_partner_performance

```sql
SELECT
  delivery_partner,
  COUNT(*) AS num_deliveries,
  ROUND(AVG(delayed) * 100, 1) AS delay_rate_percent,
  ROUND(AVG(delivery_rating), 2) AS avg_rating,
  ROUND(AVG(delivery_cost), 2) AS avg_cost
FROM workspace.delivery_logistics.delivery_logistic_clean
GROUP BY delivery_partner
HAVING COUNT(*) >= 5      -- ignore partners with very few deliveries
ORDER BY avg_rating DESC;
```

#q5_on_time_vs_delayed

```sql
SELECT
  on_time,   
  COUNT(*) AS num_deliveries,
  ROUND(AVG(distance_km), 2) AS avg_distance_km,
  ROUND(AVG(delivery_time_hours), 2) AS avg_delivery_time_hours,
  ROUND(AVG(delivery_cost), 2) AS avg_delivery_cost,
  ROUND(AVG(package_weight_kg), 2) AS avg_package_weight_kg
FROM workspace.delivery_logistics.delivery_logistic_clean
GROUP BY on_time
ORDER BY on_time;
```
